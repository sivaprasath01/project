```python
x=df[['Gender', 'Age', 'Haemoglobin', 'ESR', 'WBC', 'Neutrophil','Lymphocyte', 'Monocyte', 'Eosinophil', 'Basophil', 'RBC', 'Platelets']]
y=df[['Result']]       
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,test_size=0.3,random_state=0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
KNN=KNeighborsClassifier(n_neighbors=11)
```

```python
KNN.fit(x_train,y_train)
```

```python
KNN_pred=KNN.predict(x_test)
```

```python
KNN_pred
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python
KNN_acc=accuracy_score(y_test,KNN_pred)
```

```python
KNN_acc
```

```python
print(classification_report(y_test,KNN_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
RFC=RandomForestClassifier(n_estimators=100,max_depth=2,min_samples_leaf=2)
```

```python
RFC.fit(x_train,y_train)
```

```python
RFC_pred=RFC.predict(x_test)
```

```python
RFC_pred
```

```python
RFC_acc=accuracy_score(y_test,RFC_pred)
```

```python
RFC_acc
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR=LogisticRegression(max_iter=1000)
```

```python
LR.fit(x_train,y_train)
```

```python
LR_pred=LR.predict(x_test)
```

```python
LR_pred
```

```python
LR_acc=accuracy_score(y_test,LR_pred)
```

```python
LR_acc
```

```python
Algorithms=['KNN','RandomForest','LogisticRegression']
Accuracy=[KNN_acc,RFC_acc,LR_acc]
plt.bar(Algorithms,Accuracy,color=['#1f77b4',"#ff7f0e", "#2ca02c"])
plt.show()
```
